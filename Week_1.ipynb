# prime numbers

```python
## to find prime number from factorial we can find it
def factorial(n):
    emp_li=[]
    for i in range(1,n+1):
        if n%i==0:
            emp_li.append(i)
    return(emp_li)
```

```python
# 1st way to find
def prime(n):
    return (len(factorial)==2)
```

```python
# 2nd way is it contains 1 and same number
def prime(n):
    return (factorial(n)==[1,n])
```

```python
primeList=[]
for i in range(1,101):
    if (prime(i)):
        primeList.append(i)

print(primeList)
```

```python
def firstPrimes(n):
    (count,i,pl)=(0,1,[]) #pl -> prime list
    while (count < n):
        if prime(i):
            (count,pl)=(count+1,pl+[i])
        i+=1
    return(pl)
```

```python
print(firstPrimes(10))
```

```python
def primediffs(n):
    lastprime=2
    primediff={}
    for i in range(3,n+1):
        if prime(i):
            d=i-lastprime
            lastprime=i
            if d in primediff.keys():
                primediff[d]=primediff[d]+1
            else:
                primediff[d]=1
    return primediff
```

```python
print(primediffs(50))
```

# GCD or HCF

```python
def gcd(m,n):
    ans=[]
    for i in range(1,min(m,n)+1):
        if m%i==0 and n%i==0:
            ans.append(i)
    return(ans[-1])
```

```python
print(gcd(12,18))
```

```python
def gcd1(m,n):
    for i in range(1,min(m,n)+1):
        if m%i==0 and n%i==0:
            mrcf=i #most recent common factor
    return(mrcf)
```

```python
print(gcd1(12,18))
```
